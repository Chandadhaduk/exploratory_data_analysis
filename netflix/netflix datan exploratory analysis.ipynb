# Type of content in netflix

```python
slice=[df['type'][df["type"]=="Movie"].count(),df['type'][df["type"]=="TV Show"].count()]
label=['Movies','TV Show']
pt.pie(slice,labels=label,wedgeprops={'edgecolor':'black'},colors=['#709be0','#f2384a'],explode=[0,0.1],autopct='%1.0f%%',radius=2)
pt.show()
```

# Growth in content 

```python
y1=[df['release_year'][df['release_year']==2008][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2009][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2010][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2011][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2012][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2013][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2014][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2015][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2016][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2017][df["type"]=="Movie"].count(),
  df['release_year'][df['release_year']==2018][df["type"]=="Movie"].count()
  ]
y2=[df['release_year'][df['release_year']==2008][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2009][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2010][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2011][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2012][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2013][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2014][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2015][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2016][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2017][df["type"]=="TV Show"].count(),
  df['release_year'][df['release_year']==2018][df["type"]=="TV Show"].count()
  ]
x=[2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
pt.grid()
pt.xlabel('Years')
pt.ylabel('Number of contents')
pt.plot(x,y1,marker='.',linewidth=2,markeredgewidth=3)
pt.plot(x,y2,marker='.',linewidth=2,markeredgewidth=3)
pt.show()
```
