# Number of students in the schools

```python
b1=plt.bar(['Gabriel Pereira','Mousinho da Silveira'],[df.school[df.school=='GP'].count(),df.school[df.school=='MS'].count()],width=0.3,color='red',linewidth=1.5)
b1[0].set_hatch('/')
b1[1].set_hatch('*')
plt.title("Schools ")
plt.xlabel('School')
plt.ylabel('No. of students')
plt.show()   
```

# Relation between health and the score

```python
y=[df.health[df.health==1].count(),df.health[df.health==2].count(),df.health[df.health==3].count(),df.health[df.health==4].count(),df.health[df.health==5].count()]
x=[1,2,3,4,5]
plt.plot(x,y,color='yellow',marker='.',markerfacecolor='blue',markersize=12)
plt.title("Health VS score")
plt.xlabel('Health')
plt.ylabel('Score')
plt.show()
```
